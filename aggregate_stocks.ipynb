```python
import yfinance as yf

import pandas as pd
import pandas as pd
import os

df = pd.read_csv('constituents_csv.csv')
symbol_array = df['Symbol'].values
all_symbol = []

for symbol in symbol_array:
    print(symbol)
    contract = yf.Ticker(symbol)
    df = contract.history('3d')
    df['Symbol'] = symbol
    all_symbol.append(df)
```

```python
df = pd.concat(all_symbol)
def calculate_returns(x):
    return 100*(x['Close'].iloc[-1] - x['Open'].iloc[0])/x['Open'].iloc[0]
```

```python
df = df.reset_index()
df['dt'] = pd.to_datetime(df['Date'])
df = df.sort_values('dt')
stocks_return = df.groupby('Symbol').apply(calculate_returns)
```
